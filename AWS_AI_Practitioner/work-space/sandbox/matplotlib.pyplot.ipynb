# Explanation of `plt.figure`, `plt.scatter`, and `plt.plot` in Matplotlib

Matplotlib is a powerful Python library for data visualization.  
Here are three commonly used functions:

---

## 1. `plt.figure()`
- Creates a new **figure object** (a blank canvas) where plots will be drawn.
- You can specify the size of the figure using `figsize=(width, height)` in inches.
- Useful when you want multiple plots in separate windows or control the layout.

**Example:**
```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))  # Create a blank canvas of size 8x6 inches
```

## 2. plt.scatter()

### Description
`plt.scatter()` creates a **scatter plot**, which displays points on a 2D plane.  
It is particularly useful for visualizing the **relationship between two variables**.

### Parameters
- **x**: values on the x-axis  
- **y**: values on the y-axis  
- **color**: sets the color of the points  
- **alpha**: controls transparency (0 = fully transparent, 1 = fully opaque)  
- **label**: assigns a label for the dataset (used in legends)

``` python
# Scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y, color="blue", alpha=0.7, label="Data points")
```

## 3. plt.plot()

### Description
`plt.plot()` creates a **line plot** that connects data points in sequence.  
It is commonly used to show **trends** or **continuous relationships** between variables.

### Parameters
- **x**: values on the x-axis  
- **y**: values on the y-axis  
- **color**: sets the color of the line  
- **linewidth**: controls the thickness of the line  
- **label**: assigns a label for the dataset (used in legends)

``` python
plt.plot(x, y, color="red", linewidth=2, label="Line trend")
```

# Full Code Example

````python
import matplotlib.pyplot as plt

# Step 1: Create a figure
plt.figure(figsize=(8, 6))

# Step 2: Scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.scatter(x, y, color="blue", alpha=0.7, label="Data points")

# Step 3: Line plot
plt.plot(x, y, color="red", linewidth=2, label="Line trend")

# Add labels, title, legend, and grid
plt.title("Example of plt.figure, plt.scatter, and plt.plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
````
