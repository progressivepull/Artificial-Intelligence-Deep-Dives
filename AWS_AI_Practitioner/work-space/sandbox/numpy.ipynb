## Understanding `np.random.seed(42)`

No â€” `np.random.seed(42)` is not assigned to a variable.  
Itâ€™s a function call that sets the seed for NumPyâ€™s random number generator.

---

## ðŸ”Ž Why it works this way

- `np.random.seed(42)` initializes the random number generator with the seed `42`.
- This ensures that every time you run your code, the sequence of random numbers (like your error term) will be the same.
- You donâ€™t need to store the result because the function directly affects NumPyâ€™s internal random state.

---

```python
np.random.seed(42)  # sets the seed
print(np.random.normal(0, 1, 5))
```

# Explain with code example. np.random.normal(0, 1, 5)

# Understanding `np.random.normal(0, 1, 5)`

Letâ€™s break it down with a code example so you can see exactly what `np.random.normal(0, 1, 5)` does.

---

## ðŸ”Ž Explanation

`np.random.normal(loc, scale, size)` generates random numbers from a normal (Gaussian) distribution.

- **loc** â†’ mean of the distribution (here `0`)
- **scale** â†’ standard deviation (here `1`)
- **size** â†’ number of samples (here `5`)

So `np.random.normal(0, 1, 5)` gives you **5 random numbers from a standard normal distribution** (mean = 0, std = 1).

---


```python
# Set seed for reproducibility
np.random.seed(42)

# Generate 5 random numbers from a normal distribution
samples = np.random.normal(0, 1, 5)

print("Random samples:", samples)
```

---

## ðŸ§  Whatâ€™s happening
Each number is a draw from the **bell curve** (normal distribution) centered at `0` with a spread (standard deviation) of `1`.

---

## Changing Parameters

- `np.random.normal(10, 2, 5)`  
  â†’ mean = `10`, std dev = `2`, 5 samples

- `np.random.normal(0, 1, 1000)`  
  â†’ 1000 samples from the standard normal distribution, which you could plot as a histogram to see the bell curve.

---

```python
# Generate 1000 samples from standard normal
samples = np.random.normal(0, 1, 1000)

# Plot histogram
plt.hist(samples, bins=30, density=True, alpha=0.6, color='blue')
plt.title("Histogram of Standard Normal Samples")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
