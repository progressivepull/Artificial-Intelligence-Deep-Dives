# Simple Linear Regression Model

The equation is:



$$
y = \beta_0 + \beta_1 x + \epsilon
$$



Each term has a clear meaning:

---

## üîç Components Explained

- **$y$** ‚Üí The dependent variable (outcome)  
  *Examples:* House price, patient readmission probability, exam score  

- **$\beta_0$** ‚Üí The intercept (constant term)  
  Represents the value of $y$ when $x = 0$.  
  *Example:* If square footage is 0, the intercept might represent the baseline house price  

- **$\beta_1$** ‚Üí The slope coefficient  
  Shows how much $y$ changes when $x$ increases by 1 unit.  
  *Example:* If $\beta_1 = 200$, then each additional square foot adds \$200 to the house price  

- **$x$** ‚Üí The independent variable (predictor)  
  *Examples:* Square footage, patient age, study hours  

- **$\epsilon$** ‚Üí The error term (residual)  
  Captures everything not explained by the linear relationship between $x$ and $y$.  
  *Examples:* Noise, unmeasured factors like neighborhood quality or student motivation  

---

## ‚öñÔ∏è Intuition

Think of it like drawing the **best-fit line** through a scatterplot of data points:

- The line is defined by:
  - $\beta_0$ ‚Üí where it starts  
  - $\beta_1$ ‚Üí its tilt  

- The error term $\epsilon$ accounts for the fact that real-world data never falls perfectly on the line.


# üè† Example in Context

Suppose we want to predict house price ($y$) from square footage ($x$):



$$
\text{Price} = 50{,}000 + 200 \cdot (\text{Square Footage}) + \epsilon
$$

* **Intercept (50,000):** baseline price even for a very small house.  
* **Slope (200):** each extra square foot adds $200.  

* **Error ( $\epsilon $ ):**  accounts for factors like location, age of the house, etc.
