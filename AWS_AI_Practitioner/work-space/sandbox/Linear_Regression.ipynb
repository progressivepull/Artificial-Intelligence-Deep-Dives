# Linear Regression Model

The equation is:

$$
y = \beta_0 + \beta_1 x + \epsilon
$$

Each term has a clear meaning:

---

## üîç Components Explained

- **$y$** ‚Üí The dependent variable (outcome)  
  *Examples:* House price, patient readmission probability, exam score  

- **$\beta_0$** ‚Üí The intercept (constant term)  
  Represents the value of $y$ when $x = 0$.  
  *Example:* If square footage is 0, the intercept might represent the baseline house price  

- **$\beta_1$** ‚Üí The slope coefficient  
  Shows how much $y$ changes when $x$ increases by 1 unit.  
  *Example:* If $\beta_1 = 200$, then each additional square foot adds \$200 to the house price  

- **$x$** ‚Üí The independent variable (predictor)  
  *Examples:* Square footage, patient age, study hours  

- **$\epsilon$** ‚Üí The error term (residual)  
  Captures everything not explained by the linear relationship between $x$ and $y$.  
  *Examples:* Noise, unmeasured factors like neighborhood quality or student motivation  

---

## ‚öñÔ∏è Intuition

Think of it like drawing the **best-fit line** through a scatterplot of data points:

- The line is defined by:
  - $\beta_0$ ‚Üí where it starts  
  - $\beta_1$ ‚Üí its tilt  

- The error term $\epsilon$ accounts for the fact that real-world data never falls perfectly on the line.


# üè† Example in Context

Suppose we want to predict house price ($y$) from square footage ($x$):



$$
\text{Price} = 50{,}000 + 200 \cdot (\text{Square Footage}) + \epsilon
$$

* **Intercept (50,000):** baseline price even for a very small house.  
* **Slope (200):** each extra square foot adds $200.  

* **Error ( $\epsilon $ ):**  accounts for factors like location, age of the house, etc.


```python
import numpy as np
import matplotlib.pyplot as plt
```

# Parameters

```python
intercept = 50000   # baseline price
slope = 200         # $200 per square foot
np.random.seed(42)  # reproducibility
```

# Generate data

```python
sqft = np.linspace(500, 4000, 100)  # house sizes from 500 to 4000 sqft
error = np.random.normal(0, 20000, size=len(sqft))  # random noise (location, age, etc.)
price = intercept + slope * sqft + error
```

# Plot

```python
plt.figure(figsize=(10,6))
plt.scatter(sqft, price, color="blue", alpha=0.6, label="Observed prices (with error)")
plt.plot(sqft, intercept + slope * sqft, color="red", linewidth=2, label="Price model (no error)")

# Labels and title
plt.title("House Price vs Square Footage")
plt.xlabel("Square Footage")
plt.ylabel("Price ($)")
plt.legend()
plt.grid(True)
plt.show()
```

# Housing Price Model: Intercept, Slope, and Error

## Formula
The housing price model is defined as:

$$
y = \text{intercept} + \text{slope} \cdot x + \text{error}
$$

- **Intercept = 50,000** ‚Üí baseline price, even for a very small house.  
- **Slope = 200** ‚Üí each extra square foot adds $200.  
- **x = square footage**.  
- **Error term** ‚Üí accounts for unpredictable factors like location, age, or condition.

---

## Examples
- At **0 sqft**:  
  
$$
y = 50{,}000
$$

- At **750 sqft**:  
  
$$
y = 50{,}000 + 200 \cdot 750 = 200{,}000
$$

- At **1,000 sqft**:  
  
$$
y = 50{,}000 + 200 \cdot 1000 = 250{,}000
$$

---

## Key Insight
- The **intercept** is always **50,000**.  
- The larger values (like 200,000 or 250,000) come from plugging in specific square footage values into the formula.  
- The error term shifts these values up or down to reflect real-world variability.
